```python
import pandas as pd
from sqlalchemy import create_engine, inspect

engine = create_engine('postgresql://localhost/customer_intelligence')
inspector = inspect(engine)

print("Veritabanındaki tablolar:", inspector.get_table_names())
```

```python
for table in ['customers', 'predictions', 'models', 'predictionlogs']:
    print(f"\n=== {table.upper()} ===")
    columns = inspector.get_columns(table)
    for col in columns:
        print(f"  {col['name']:25s} {str(col['type']):20s} nullable={col['nullable']}")
```

```python
for table in ['customers', 'predictions', 'models', 'predictionlogs']:
    count = pd.read_sql(f'SELECT COUNT(*) as adet FROM {table}', engine)
    print(f"{table}: {count['adet'].iloc[0]} kayıt")
```

```python
customers_check = pd.read_sql('SELECT * FROM customers', engine)

print("Eksik değerler:")
print(customers_check.isnull().sum())
print()
print("Tekrarlayan CustomerId var mı:", customers_check['customerid'].duplicated().sum())
print()
print("Age aralığı:", customers_check['age'].min(), "-", customers_check['age'].max())
print("MonthlyCharge aralığı:", customers_check['monthlycharge'].min(), "-", customers_check['monthlycharge'].max())
print()
print("Gender değerleri:", customers_check['gender'].unique())
print("ContractType değerleri:", customers_check['contracttype'].unique())
print("InternetService değerleri:", customers_check['internetservice'].unique())
```

```python
predictions_check = pd.read_sql('SELECT * FROM predictions', engine)

print("Eksik değerler:")
print(predictions_check.isnull().sum())
print()
print("ChurnProbability aralığı:", predictions_check['churnprobability'].min(), "-", predictions_check['churnprobability'].max())
print()
print("RiskLevel dağılımı:")
print(predictions_check['risklevel'].value_counts())
print()
print("SegmentLabel dağılımı:")
print(predictions_check['segmentlabel'].value_counts())
```

```python
# Predictions tablosundaki her CustomerId, Customers tablosunda gerçekten var mı?
orphan_check = pd.read_sql("""
    SELECT p.customerid 
    FROM predictions p 
    LEFT JOIN customers c ON p.customerid = c.customerid 
    WHERE c.customerid IS NULL
""", engine)

print("Customers tablosunda karşılığı olmayan Prediction kaydı sayısı:", len(orphan_check))
```
