```python
import pandas as pd
import numpy as np
from faker import Faker

df = pd.read_csv('ibm_telco.csv')
print(df.shape)
df.head()
```

```python
fake = Faker('tr_TR')

# Name (cinsiyete göre)
def generate_name(gender):
    if gender == 'Male':
        first = fake.first_name_male()
        last = fake.last_name_male()
    else:
        first = fake.first_name_female()
        last = fake.last_name_female()
    return f"{first} {last}"

df['Name'] = df['gender'].apply(generate_name)

# Age (18-80 arası mantıklı bir dağılım)
np.random.seed(42)
df['Age'] = np.random.randint(18, 81, size=len(df))

# SupportTicketCount (TechSupport'a bağlı Poisson)
df['SupportTicketCount'] = np.where(
    df['TechSupport'] == 'No',
    np.random.poisson(lam=3, size=len(df)),
    np.random.poisson(lam=1, size=len(df))
)

df[['customerID', 'gender', 'Name', 'Age', 'SupportTicketCount']].head(10)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(df['TotalCharges'].isnull().sum())
```

```python
df.to_csv('telecom_ibm_processed.csv', index=False)
```
